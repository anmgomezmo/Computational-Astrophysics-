```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
T = 365.25635
A = 1 # in UA = 1.496X10^8 km
e = 0.0167
omega = 2*np.pi/T 
omega 
```

```python
def ecc_anomaly(E,t):
    return E - omega*t - e*np.sin(E)
```

```python
t = 91 
E = np.linspace(-1e1,1e1,10000)
f = ecc_anomaly(E,t)
plt.figure(figsize=(10,7))
plt.plot(E,f,label='Ecc anomaly function')
plt.xlabel('E')
plt.ylabel('f(E)')
plt.xlim(0,2)
plt.ylim(-2,2)
plt.grid()
plt.legend()
plt.show()
```

```python
# Find the initial and final points of the restricted interval of 
# three points to interpolate.
for i in range(len(E)-1):
    if E[i]>=-1:
        a=i
        break

# The interval will include three points:
# x[a], x[a+1], x[a+2]
# Here we define the arrays that enter as arguments
# for the interpolated polynomial
x_int =[E[a], E[a+1], E[a+2]]
f_int =[f[a], f[a+1], f[a+2]]

print('a = ', a, ': x_a = ', E[a], ': f(x_a) = ', f[a])
```

```python
# 2-order Interpolated Polynomial 
def p2(x):
    pol = (((x-x_int[1])*(x-x_int[2]))/((x_int[0]-x_int[1])*(x_int[0]-x_int[2])))*f_int[0] +\
          (((x-x_int[0])*(x-x_int[2]))/((x_int[1]-x_int[0])*(x_int[1]-x_int[2])))*f_int[1] +\
          (((x-x_int[0])*(x-x_int[1]))/((x_int[2]-x_int[0])*(x_int[2]-x_int[1])))*f_int[2]
    return pol


# Newton-Raphson Method. We begin with the value at point x[a]
root0 = E[a]-1e-5
root = E[a]
iterations = 0

while np.abs(p2(root))>1e-10:
    iterations += 1
    new_root = root - p2(root)*(root - root0)/(p2(root) - p2(root0))
    root0, root = root, new_root

print(f'The root is x_r = {root:.8f}')
print(f'for which the interpolated function gives f(x_r) =  {p2(root):.8e}')
print(f'Number of iterations = {iterations}')
```

```python
times = [91,182,273]
boundary = [1,2,4]
index = 0
E_s = np.zeros(3)
f_E = np.zeros(3)
for t,j in zip(times,boundary):
    f = ecc_anomaly(E,t)
    for i in range(len(E)-1):
        if E[i]>=j:
            a=i
            break
    
    root0 = E[a]-1e-5
    root = E[a]    
    x_int =[E[a], E[a+1], E[a+2]]
    f_int =[f[a], f[a+1], f[a+2]]
    while np.abs(p2(root))>1e-10:
        iterations += 1
        new_root = root - p2(root)*(root - root0)/(p2(root) - p2(root0))
        root0, root = root, new_root

    print(f'The root is E = {root:.8f}')
    print(f'for which the interpolated function gives f(E) =  {p2(root):.8e}')
    print(f'Number of iterations = {iterations}')
    
    E_s[index] = root
    f_E[index] = p2(root)
    index += 1
```

```python
plt.subplots(nrows=1,ncols=3,figsize=(16,5),sharex=True,sharey=True)
for i,t in zip(range(0,3),times):
    plt.subplot(1,3,i+1)
    plt.plot(E, ecc_anomaly(E,t), color='red')
    plt.scatter(E_s[i], f_E[i], color='blue', marker='*',s=150)
    plt.vlines(E_s[i], -7, f_E[i], linestyle='dashed')
    plt.hlines(f_E[i],0, E_s[i], linestyle='dashed')
    plt.title('Root')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$f(x)$')
    plt.xlim(0,6)
    plt.ylim(-4,2)
    plt.grid()
plt.show()
```

```python
B = A*np.sqrt(1-e**2)
B
```

```python
x = np.zeros((1,3))
y = np.zeros((1,3))

for i in range(0,3):
    x[0,i] = A*np.cos(E_s[i])
    y[0,i] = B*np.sin(E_s[i])

print(x,'\n',y)
```

```python
from matplotlib.patches import Ellipse
e = Ellipse((0,0),2*A,2*B,angle=0,linewidth=1,fill=False)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111,aspect='auto')
ax.add_patch(e)
plt.plot(x,y,'*r',label='Coordinates',markersize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
