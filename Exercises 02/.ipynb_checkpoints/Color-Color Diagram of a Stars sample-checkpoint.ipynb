```python
g_r = dcc['g']-dcc['r']
r_i = dcc['r']-dcc['i']
u_g = dcc['u']-dcc['g']
i_z = dcc['i']-dcc['z']
print(g_r)
```

```python
plt.figure(figsize=(5,5))
plt.plot(g_r,r_i,'.k')
plt.xlabel('g-r')
plt.ylabel('r-i')
plt.show()
```


Como la magnitud decrese cuando la intensidad aumenta, la magnitud en los picos de las curvas es negativa y grande en magnitud, mientras que en las caidas la magnitud aumenta y puede ser cercana o mayor que cero. \
Por lo tanto siguiendo los espectros de las estrellas como el de la figura anterior, si la diferencia entre g-r y r-i es pequea la estrella presenta una temperatura relativamente alta (caliente), mientras que si la diferencia entre g-r y r-i es grande la estrella presenta una temperatura relativamente baja (fria). \
Por ello las estrellas en la parte inferior izquierda son calientes mientras que las de la parte superior derecha son frias. 

```python
plt.figure(figsize=(5,5))
plt.plot(u_g,g_r,'.k')
plt.xlabel('u-g')
plt.ylabel('g-r')
plt.show()
```

La linea de tendencia que las estrellas parecen seguir se rompe aproximadamente en 2.1 de u-g y en 1.0 en g-r, con esto se puede interpretar que las estrellas con coordenadas menores al punto de quiebre anterior son fuentes termicas que transportan calor a traves de la radiacion emitida en sus respectivos picos del espectro, mientras que las demas liberan menos intensidad y como sus picos son mas aplanados liberan radiacion pareja en mas longitudes de onda que las estrellas calientes, por ello estas estrellas con menor intensidad menor radiacion y por lo tanto menos calor liberan al espacio exterior. \

Siguiendo la utilizacion del espectro recibido de las estrellas, como el de la imagen arriba, las estrellas amontonadas sobre el sector en el cual u-g no varia mucho y g-r varia (parte superior derecha), presentan una pequea variacion en u-g lo que implica que la parte inicial de su espectro son muy parecidos, pero al variar g-r, implica que el pico del espectro de estas estrellas crece tal que en rojo aumenta la intensidad y por ello el aumento de g-r, por lo cual estas estrellas emiten principalmente en el color rojo y por su espectro presentan temperaturas superficiales bajas, y estas son las conocidas enenas rojas, las cuales son las mas abundantes del universo.
