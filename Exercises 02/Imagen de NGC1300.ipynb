```python
from astropy.io import fits
from astropy.visualization import make_lupton_rgb
from matplotlib import pyplot as plt 
```

```python
ngc_b = fits.open("ngc1300B.fits")
ngc_i = fits.open("ngc1300I.fits")
ngc_k = fits.open("ngc1300K.fits")
```

```python
b = ngc_b[0].data
i = ngc_i[0].data
k = ngc_k[0].data
```

```python
rgb_default = make_lupton_rgb(k,i,b)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rgb_default,origin='lower')
plt.show()
```

```python
rgb_default = make_lupton_rgb(k,i,b,stretch=2,Q=30)
plt.figure(figsize=(10,10))
plt.imshow(rgb_default,origin='lower')
plt.show()
```
