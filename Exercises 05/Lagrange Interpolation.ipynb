```python
import numpy as np 
import matplotlib.pyplot as plt 
import LagrangeInterpolation as lag
```

```python
def f(x):
    return 1.0/(25*x*x+1)
```

```python
N = 1000
x = np.linspace(-1,1,N)
func = f(x)
plt.figure(figsize=(10,6))
plt.plot(x,func,'-r',label='Original function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Define an array of points of variable x and its values of f 
N = 7
x = np.linspace(-1,1,N)
func = f(x)
# Interpolate the last points with a polynomial
n = 6
plt.figure(figsize=(10,6))
plt.scatter(x,func,label='Original function')

for i in range(N-n):
    xi = x[i:i+n+1]
    fi = func[i:i+n+1]
    x_interval_1 = np.linspace(x[i],x[i+1],3*n)
    y_interval_1 = lag.p(x_interval_1,xi,fi)
    plt.plot(x_interval_1, y_interval_1,'r')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
