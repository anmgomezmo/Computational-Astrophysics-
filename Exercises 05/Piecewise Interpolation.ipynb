```python
import numpy as np 
import matplotlib.pyplot as plt 
import LagrangeInterpolation as lag
```

```python
def f(x):
    return 1.0/(25*x*x+1)
```

```python
def EN2(m,p,f):
    summ = 0
    for i in range(m):
        summ += ((p[i]-f[i])/f[i])**2.0
    
    EN2 = np.sqrt(summ)/m
    return EN2
```

```python
# Define an array of points of variable x and its values of f 
N = 50
x = np.linspace(-1,1,N)
func = f(x)
# Interpolate the last points with a polynomial
n = 1
plt.figure(figsize=(10,6))
plt.scatter(x,func,label='Original function')

p = np.zeros(1)
for i in range(N-n):
    xi = x[i:i+n+1]
    fi = func[i:i+n+1]
    x_interval_1 = np.linspace(x[i],x[i+1],3*n)
    y_interval_1 = lag.p(x_interval_1,xi,fi)
    p = np.concatenate((p,y_interval_1[0]),axis=None)
    plt.plot(x_interval_1, y_interval_1,'r')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print(p.shape,func.shape)
print('EN2: %.4f' % EN2(N-n,p[:],func[:94]))
```
