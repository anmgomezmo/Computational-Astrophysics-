```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import LagrangeInterpolation as lag
import HermiteInterpolation as her
```

```python
data = pd.read_csv('cepheids_data.csv')
data
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data['Time [days]'],data['Apparent Magnitude'])
plt.xlabel('Time [days]')
plt.ylabel('Apparent Magnitude')
plt.show()
```

```python
# Define an array of points of variable x and its values of f 
x = data['Time [days]']
y = data['Apparent Magnitude']
N = len(x)
# Interpolate the last points with a polynomial
n = 8
plt.figure(figsize=(10,7))
plt.scatter(x,y,label='Original function')

for i in range(N-n):
    xi = x[i:i+n+1]
    fi = y[i:i+n+1]
    x_interval = np.linspace(x[i],x[i+1],3*n)
    y_interval = lag.p(x_interval,xi,fi)
    plt.plot(x_interval, y_interval,'r',label='Polynomial Interpolated')

plt.xlabel('Time [days]')
plt.ylabel('Apparent Magnitude')
plt.legend()
plt.show()
```

```python
def linearInterpolation(x1, x2, f1, f2, x):
    p1 = f1 + ((f2-f1)/(x2-x1))*(x-x1)
    return p1

def quadraticInterpolation(x1, x2, x3, f1, f2, f3, x):
    p2 = (((x-x2)*(x-x3))/((x1-x2)*(x1-x3)))*f1 + (((x-x1)*(x-x3))/((x2-x1)*(x2-x3)))*f2 +\
         (((x-x1)*(x-x2))/((x3-x1)*(x3-x2)))*f3
    return p2

plt.figure(figsize=(10,7))
plt.scatter(x, y, color='black')

for i in range(N-1):
    x_interval = np.linspace(x[i],x[i+1],3) 
    # Note that the number 3 in thie above line indeicates the number of 
    # points interpolated in each interval !
    # (including the extreme points of the interval)
    y_interval = linearInterpolation(x[i], x[i+1], y[i], y[i+1], x_interval)
    plt.plot(x_interval, y_interval,'r')
    
for i in range(N-2):
    x_interval = np.linspace(x[i],x[i+1],6) # 6 interpolate points in each interval
    y_interval = quadraticInterpolation(x[i], x[i+1], x[i+2], y[i], y[i+1], y[i+2], x_interval)
    plt.plot(x_interval, y_interval,'g')
    
plt.title(r'Linear and Cuadratic Polynomial Piecewise Interpolation')
plt.xlabel('Time [days]')
plt.ylabel('Apparent Magnitude')
plt.show()
```

```python
def Derivative(x, f):
    '''
    ------------------------------------------
    Derivative(x, f)
    ------------------------------------------
    This function returns the numerical 
    derivative of a discretely-sample function 
    using one-side derivatives in the extreme 
    points of the interval and second order 
    accurate derivative in the middle points.
    The data points may be evenly or unevenly
    spaced.
    ------------------------------------------
    '''
    # Number of points
    N = len(x)
    dfdx = np.zeros([N, 2])
    dfdx[:,0] = x
    
    # Derivative at the extreme points
    dfdx[0,1] = (f[1] - f[0])/(x[1] - x[0])
    dfdx[N-1,1] = (f[N-1] - f[N-2])/(x[N-1] - x[N-2])
    
    #Derivative at the middle points
    for i in range(1,N-1):
        h1 = x[i] - x[i-1]
        h2 = x[i+1] - x[i]
        dfdx[i,1] = h1*f[i+1]/(h2*(h1+h2)) - (h1-h2)*f[i]/(h1*h2) - h2*f[i-1]/(h1*(h1+h2))
    
    return dfdx
```

```python
# Calling the derivative function and chosing only the second column
dfdx = Derivative(x,y)[:,1]

plt.figure(figsize=(10,7))
plt.title(f'Cubic Hermite Polynomial Piecewise Interpolation')
plt.scatter(x, y, color='black')

# Piecewise Hermite Interpolation Loop
for i in range(N-1):
    xi = x[i:i+2]
    fi = y[i:i+2]
    dfidx = dfdx[i:i+2]
    x_interval = np.linspace(x[i],x[i+1],4)
    y_interval = her.H3(x_interval, xi, fi, dfidx)
    plt.plot(x_interval, y_interval,'r')

plt.xlabel('Time [days]')
plt.ylabel('Apparent Magnitude')
plt.show()
```
