```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def f(x,s,m):
    return np.exp(-0.5*((x-m)/s)**2.0)/(np.sqrt(2*np.pi)*s)
```

```python
def Derivative(x,h,s,m,forward=False,backward=False,central=False):
    if forward:
        return (f(x+h,s,m)-f(x,s,m))/h
    if backward:
        return (f(x,s,m)-f(x-h,s,m))/h
    if central:
        return (f(x+h,s,m)-f(x-h,s,m))/(2*h)
```

```python
a = -10.0
b = 10.0
s = 1.0
m = 0
N = 1000
h = (b-a)/N
x = np.linspace(a,b,N)
func = f(x,s,m)
deriv = Derivative(x,h,s,m,forward=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,func,label='Gaussian function')
plt.plot(x,deriv,label='Derivate')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
s = 2.0
m = 1.5
N = 1000
a = -10.0*s+m
b = 10.0*s+m
h = (b-a)/N
x = np.linspace(a,b,N)
func = f(x,s,m)
deriv = Derivative(x,h,s,m,central=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,func,label='Gaussian function')
plt.plot(x,deriv,label='Derivate')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def deriv_analytic(x,s,m):
    return ((m-x)*np.exp(-0.5*((x-m)/s)**2.0))/(np.sqrt(2*np.pi)*s**3.0)
```

```python
s = 2.0
m = 1.5
N = 1000
a = -10.0*s+m
b = 10.0*s+m
h = (b-a)/N
x = np.linspace(a,b,N)
deriv = Derivative(x,h,s,m,central=True)
analytic = deriv_analytic(x,s,m)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,deriv,label='Derivate')
plt.plot(x,analytic,label='Analaytic Derivate')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
