```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
h_p = 6.62606876e-34
c = 3e8
k = 1.3806503e-23
```

```python
def f(wl,T):
     return ((2*h_p*c*c)/(wl**5.0))/(np.exp((h_p*c)/(wl*k*T))-1.0)
```

```python
def Derivate(wl,T,h,forward=False,backward=False,central=False):
    if forward:
        return (f(wl+h,T)-f(wl,T))/h
    if backward:
        return (f(wl,T)-f(wl-h,T))/h
    if central:
        return (f(wl+h,T)-f(wl-h,T))/(2*h)
```

```python
a = 1e-7
b = 3e-6
T = 6000
N = 1000
h = (b-a)/N
x = np.linspace(a,b,N)
func = f(x,T)
deriv = Derivate(x,T,h,central=True)
```

```python
plt.subplots(nrows=1,ncols=2,figsize=(15,6),sharex=True,sharey=True)
plt.subplot(121)
plt.plot(x,func,label='Planck function')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$B(\lambda)$')
plt.legend()
plt.grid()
plt.subplot(122)
plt.plot(x,deriv,label='Derivate')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$B(\lambda)$')
plt.legend()
plt.grid()
plt.show()
```
