```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def f(x,s,m):
     return np.exp(-0.5*((x-m)/s)**2.0)/(np.sqrt(2*np.pi)*s)
```

```python
def Derivative(x,h,s,m,forward=False,backward=False,central=False):
    if forward:
        return (f(x+h,s,m)-f(x,s,m))/h
    if backward:
        return (f(x,s,m)-f(x-h,s,m))/h
    if central:
        return (f(x+h,s,m)-f(x-h,s,m))/(2*h)
```

```python
def second_Derivative(x,h,s,m,derivative,forward=False,backward=False):
    if forward:
        return (f(x+h,s,m)-f(x,s,m)-derivative*h)/(0.5*h**2.0)
    if backward:
        return (f(x,s,m)-f(x-h,s,m)+derivative*h)/(0.5*h**2.0)
```

```python
s = 2.0
m = 1.5
N = 1000
a = -10.0*s+m
b = 10.0*s+m
h = (b-a)/N
x = np.linspace(a,b,N)
func = f(x,s,m)
deriv = Derivative(x,h,s,m,central=True)
second_deriv = second_Derivative(x,h,s,m,deriv,forward=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,func,label='Gaussian function')
plt.plot(x,deriv,label='Derivate')
plt.plot(x,second_deriv,label='Second Derivate')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def second_derivade_f_analytic(x,s,m):
     return np.exp(-0.5*((x-m)/s)**2.0)*((m-x)**2.0-s**2.0)/(np.sqrt(2*np.pi)*s**5.0)
```

```python
s = 2.0
m = 1.5
N = 1000
a = -10.0*s+m
b = 10.0*s+m
h = (b-a)/N
x = np.linspace(a,b,N)
deriv = Derivative(x,h,s,m,central=True)
second_deriv = second_Derivative(x,h,s,m,deriv,forward=True)
second_deriv_analytic = second_derivade_f_analytic(x,s,m)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,second_deriv,label='Second Derivate')
plt.plot(x,second_deriv_analytic,label='Second Derivate analytic')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
