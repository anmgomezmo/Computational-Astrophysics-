```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def PlanckFunc(T,l):
    c = 3e8
    k = 1.3806503e-23 
    h = 6.62606876e-34
    B = ((2*h*c*c)/(l**5.0))/(np.exp((h*c)/(l*k*T))-1.0)
    return B 
```

```python
def PlanckFunc_Temp(T):
    n = 1000
    l = np.linspace(1e-7,3e-6,n)
    B = PlanckFunc(T,l)
    plt.figure(figsize=(15,8))
    plt.plot(l,B,label='T=%i k' % T)
    #plt.xlabel(r'$\lambda$')
    #plt.xlabel('Wavelenght')
    plt.ylabel(r'$B(\lambda)$')
    plt.legend()
    plt.savefig('Planck_function_T%i.pdf' % T,dpi=600)
    plt.show()
```

```python
PlanckFunc_Temp(5000)
PlanckFunc_Temp(6000)
PlanckFunc_Temp(7000)
PlanckFunc_Temp(8000)
```
