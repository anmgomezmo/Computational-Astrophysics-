```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def GaussFunc(m,s):
    N = 1000
    x = np.linspace(-10,10,N)
    f = np.exp(-0.5*((x-m)/s)**2.0)/(np.sqrt(2*np.pi)*s)
    plt.figure(figsize=(15,8))
    plt.plot(x,f,'-r',label='Gaussian')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.savefig('Gauss_Function.jpg',dpi=600)
    plt.savefig('Gauss_Function.pdf',dpi=600)
    plt.show()
```

```python
GaussFunc(0,1)
```
