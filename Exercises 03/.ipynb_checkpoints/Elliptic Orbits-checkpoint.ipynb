```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def EllipticOrbit(phi,e,a):
    r = a*(1-e*e)/(1+e*np.cos(phi))
    return r
```

```python
e = 0.5
a = 1.0
N = 1000
phi = np.linspace(0,2*np.pi,N)
r = EllipticOrbit(phi,e,a)
```

```python
x = r*np.cos(phi)
y = r*np.sin(phi)
x.shape
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,y,'-r',label='Elliptic orbit')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
